```python

columns = ['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN']
for col in columns:
    new_train[col] = new_train[col].astype(int)
print(new_train)
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


y = new_train.median_house_value 
x =  new_train.drop('median_house_value',axis='columns')


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


model = LinearRegression()


model.fit(x_train, y_train)


y_pred = model.predict(x_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


print("Mean Squared Error:", mse)
print("R^2 Score:", r2)

coefficients = pd.DataFrame({'Feature': x.columns, 'Coefficient': model.coef_})
print(coefficients)

```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(x_train , y_train)
```

```python
forest.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model_gnb = GaussianNB()
model_gnb.fit(x_train, y_train)
model_gnb.score(x_test,y_test)
```

```python
def predict_median_house_value(features, model):

    input_data = pd.DataFrame([features])
    
    
    predicted_value = model.predict(input_data)
    
    return predicted_value[0]
```

```python
example_features = {
    'longitude': -122.23,
    'latitude': 37.88,
    'housing_median_age': 41,
    'total_rooms': 880,
    'total_bedrooms': 129,
    'population': 322,
    'households': 126,
    'median_income': 8.3252,
     'bedroom_ratio': 0.1466,  
    'household_rooms': 6.9841,
    '<1H OCEAN': 1,
    'INLAND': 0,
    'ISLAND': 0,
    'NEAR BAY': 0,
    'NEAR OCEAN': 0,
    
}

predicted_value = predict_median_house_value(example_features, forest)
print("Predicted Median House Value:", predicted_value)
```
